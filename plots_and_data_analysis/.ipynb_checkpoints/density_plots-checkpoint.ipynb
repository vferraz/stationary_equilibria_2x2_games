```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
neia = pd.read_csv('data_equilibria/sg_output_NEIA.csv')
ibeia = pd.read_csv('data_equilibria/sg_output_IBEIA.csv')
ibeia2 = pd.read_csv('data_equilibria/sg_output_IBE2IA.csv')
ibegamma = pd.read_csv('data_equilibria/sg_output_IBEGAMMA.csv')
qre = pd.read_csv('data_equilibria/qre_lambda_out.csv')
```

```python
ibegammaia = pd.read_csv('sg_output_IBEGAMMAIA.csv') # this file is heavy and it takes a while to load
```

```python
#neia = neia.loc[neia['q'] != 2]
neia = neia.loc[neia['framed'] == 1]
neia = neia.loc[neia['q'] != 2]

ibeia = ibeia.loc[ibeia['q'] != 2]
ibeia = ibeia.loc[ibeia['framed'] == 1]

ibeia2 = ibeia2.loc[ibeia2['q'] != 2]
ibeia2 = ibeia2.loc[ibeia2['framed'] == 1]

ibegamma = ibegamma.loc[ibegamma['q'] != 2]
ibegamma = ibegamma.loc[ibegamma['framed'] == 1]
```

```python
ibegammaia = ibegammaia.loc[ibegammaia['q'] != 2]
ibegammaia = ibegammaia.loc[ibegammaia['framed'] == 1]
```

```python
fr =[0.461,0.136]
uf =[0.341,0.108]

kwargs  =   {
             'linewidth':0.02, # line width of spot
             'alpha':0.3
            }
s=1

c_line='grey'
height_c = 3.6
```

```python
g = sns.JointGrid(data=neia, x="PU", y="QL",height=height_c)
g.plot_joint(sns.scatterplot, s=s,**kwargs)
g.ax_joint.scatter(fr[0],fr[1], c='red', marker='x', s=75, label='framed_mean')
g.ax_joint.scatter(uf[0],uf[1], c='blue', marker='+', s=125, label='framed_mean')
g.plot_marginals(sns.kdeplot, fill=True)
g.ax_marg_x.set_xlim(0, 1)
g.ax_marg_y.set_ylim(0, 1)
g.fig.suptitle("NE‚ó¶IA(Œ±,Œ≤)")
plt.savefig('neia.png', format='png', dpi=1200, transparent=True)
```

```python
qre = qre.rename(columns={"p1_s1": "PU", "p2_s1": "QL"})
g = sns.JointGrid(data=qre, x="PU", y="QL",height=height_c)
g.plot_joint(sns.scatterplot, s=s,**kwargs)
g.ax_joint.scatter(fr[0],fr[1], c='red', marker='x', s=75, label='framed_mean')
g.ax_joint.scatter(uf[0],uf[1], c='blue', marker='+', s=125, label='framed_mean')
g.plot_marginals(sns.kdeplot, fill=True)
g.fig.suptitle("QREŒª")
```
